```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU, Bidirectional
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

maxlen=100
x_train = pad_sequences(x_train, maxlen=maxlen, padding='post', truncating='post')
x_test = pad_sequences(x_test, maxlen=maxlen, padding='post', truncating='post')
```

```python
# simple RNN
model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(maxlen,)),
    SimpleRNN(5),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
# bidirectional RNN
model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(maxlen,)),
    Bidirectional(SimpleRNN(5)),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
from IPython.core import history
history = model.fit(x_train,y_train,epochs=5,batch_size=32, validation_split=0.2)
```

```python
# bidirectional lstm

model = Sequential([
    Embedding(10000,32,input_shape=(100,)),
    Bidirectional(LSTM(5)),
    Dense(1,activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from IPython.core import history
history = model.fit(x_train,y_train,epochs=5,batch_size=32, validation_split=0.2)
```

```python
# bidirectional GRU

model = Sequential([
    Embedding(10000,32,input_shape=(100,)),
    Bidirectional(GRU(5)),
    Dense(1,activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from IPython.core import history
history = model.fit(x_train,y_train,epochs=5,batch_size=32, validation_split=0.2)
```
